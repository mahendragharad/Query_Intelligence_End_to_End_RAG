#### Text Splitting from Documents- 1. RecursiveCharacter Text Splitters
This text splitter is the recommended one for generic text. It is parameterized by a list of characters. It tries to split on them in order until the chunks are small enough. The default list is ["\n\n", "\n", " ", ""]. This has the effect of trying to keep all paragraphs (and then sentences, and then words) together as long as possible, as those would generically seem to be the strongest semantically related pieces of text.

- How the text is split: by list of characters.
- How the chunk size is measured: by number of characters.


```python
# Let's perform text splitting using RecursiveTextSplitter 

# 1. Read the pdf file 
from pathlib import Path 
from langchain_community.document_loaders import PyMuPDFLoader

BASE_DIR = Path.cwd().resolve().parent
file_path = BASE_DIR / "config" / "data" / "processed" / "AttentionYou.pdf"

loader = PyMuPDFLoader(str(file_path))
text_document = loader.load()

# 2. Let's Perform the chunking 
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
splitted_document = text_splitter.split_documents(text_document)
splitted_document
```

### Advance One With RecursiveTextSplitter


```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# for better chunking add some separators
splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=800,
    chunk_overlap=120,
    separators=[
        "\n# ",
        "\n## ",
        "\n### ",
        "\n\n",
        "\n",
        ". ",
        "? ",
        "! ",
        " "
    ]
)

# let's break it down 
chunks = splitter.split_documents(text_document)

# Let's see the optimize one 
chunks
```

### Let's See with the Text file also 

```python
# Let's perform text splitting using RecursiveTextSplitter 

# 1. Read the pdf file 
from pathlib import Path 
from langchain_community.document_loaders import PyMuPDFLoader

BASE_DIR = Path.cwd().resolve().parent
file_path2 = BASE_DIR / "config" / "data" / "raw" / "demo.txt"

loader2 = PyMuPDFLoader(str(file_path2))
text_document2 = loader2.load()

# 2. Let's Perform the chunking 
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter2 = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
text_splitted = text_splitter2.split_documents(text_document2)
text_splitted
```

### 2. Let's See About the CharacterTextSplitter

```python
from langchain_text_splitters import CharacterTextSplitter
char_splitter = CharacterTextSplitter(separator="\n\n",chunk_size=100, chunk_overlap=20)

char_data = ""

with open(file_path2, 'r') as file :
    char_data = file.read()

splitted_char = char_splitter.create_documents([char_data])
splitted_char
```

### 3. Let's See About HTML Text Splitter 

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

url = "https://plato.stanford.edu/entries/goedel/"

headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
    ("h4", "Header 4"),
]
html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
html_header_splits = html_splitter.split_text_from_url(url)
html_header_splits
```

### Let's See about the JSON-TextSplitter

```python
# Let's go through it 
import json
import requests

json_data = requests.get("https://api.smith.langchain.com/openapi.json").json()
```

```python
# Let's see our json data 
json_data
```

```python
# Let's Perform chunking 
from langchain_text_splitters import RecursiveJsonSplitter

json_splitter = RecursiveJsonSplitter(max_chunk_size=100)
json_splitted = json_splitter.split_json(json_data)
json_splitted
```

```python
## The splitter can also output documents
docs=json_splitter.create_documents(texts=[json_data])
for doc in docs[:3]:
    print(doc)
```

```python
for chunk in json_splitted[:3] :
    print(chunk)
```

```python
# using split text 
split_text = json_splitter.split_text(json_data)
for chunk in split_text[:5] :
    print(chunk)
```
